```python
generic_cols = [
        "z",
        "zErr",
        "ra",
        "dec",
        "plate",
        "mjd",
        "expMag_r",
        "deVMag_r",
        "cModelMag_r",
        "type"
    ]
```

```python
for col in generic_cols:
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```
