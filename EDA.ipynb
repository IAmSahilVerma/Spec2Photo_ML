```python
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

def clean_galaxy_data(df):
    # --- Column groups ---
    spectro_cols = [
        "h_alpha_flux", "oiii_5007_flux", "nii_6584_flux", "sii_6717_flux",
        "lick_mgb", "lick_fe5335", "d4000", "velocityDisp"
    ]
    photometry_cols = [
        "modelMag_u", "modelMag_r", "modelMag_z", "fracDev_r", "petroRad_r",
        "extinction_u", "extinction_r", "extinction_z"
    ]
    generic_cols = ["z", "zErr"]
    target_col = ["logM_star"]

    # --- Step 1: Log-transform emission line fluxes ---
    for col in ["h_alpha_flux", "oiii_5007_flux", "nii_6584_flux", "sii_6717_flux"]:
        df[f"log_{col}"] = np.log1p(df[col])
    
    # --- Step 2: Physical filtering ---
    df = df[df["z"].between(0.001, 1.0)]
    df = df[df["zErr"] < 0.01]
    df = df[df["petroRad_r"] > 0.5]
    df = df[df["fracDev_r"].between(0, 1)]
    df = df[df["extinction_u"] < 1.5]

    # --- Step 3: Statistical filtering ---
    def iqr_filter(col):
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        return df[col].between(Q1 - 1.5 * IQR, Q3 + 1.5 * IQR)

    for col in ["logM_star", "velocityDisp", "d4000"]:
        df = df[iqr_filter(col)]

    # --- Step 4: Multivariate outlier detection ---
    feature_cols = [f"log_{col}" for col in ["h_alpha_flux", "oiii_5007_flux", "nii_6584_flux", "sii_6717_flux"]] \
                   + ["lick_mgb", "lick_fe5335", "d4000", "velocityDisp"] \
                   + photometry_cols + generic_cols

    df_features = df[feature_cols].dropna()
    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(df_features)

    iso = IsolationForest(contamination=0.02, random_state=42)
    outlier_flags = iso.fit_predict(features_scaled)

    df_clean = df.loc[df_features.index[outlier_flags == 1]].copy()

    # --- Optional: PCA visualization ---
    pca = PCA(n_components=2)
    pca_result = pca.fit_transform(features_scaled)
    plt.figure(figsize=(8, 6))
    plt.scatter(pca_result[:, 0], pca_result[:, 1], c=(outlier_flags == -1), cmap='coolwarm', alpha=0.6)
    plt.title("PCA of Galaxy Features with Outliers Highlighted")
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.grid(True)
    plt.show()

    return df_clean
```

🧠 Step-by-Step Explanation of the Outlier Removal Pipeline
1. 🔬 Log-Transform Emission Line Fluxes
np.log1p(df[col])


✅ Why:
- Emission line fluxes (e.g. Hα, [OIII]) often span several orders of magnitude.
- Log-transforming stabilizes variance and reduces skewness, making statistical methods like Z-score or Isolation Forest more effective.
- log1p handles zero values gracefully, avoiding log(0) errors.

2. 🌌 Physical Filtering
df[df["z"].between(0.001, 1.0)]
df[df["zErr"] < 0.01]
df[df["petroRad_r"] > 0.5]
df[df["fracDev_r"].between(0, 1)]
df[df["extinction_u"] < 1.5]


✅ Why:
- Redshift bounds: Removes non-physical or misclassified sources (e.g. stars or artifacts).
- Redshift error: High uncertainty can distort derived quantities like luminosity or mass.
- Petrosian radius: Very small radii may indicate unresolved or blended sources.
- FracDev: Should lie between 0 and 1; values outside this range suggest catalog errors.
- Extinction: Extremely high extinction values are rare and often due to poor photometry or misclassification.

3. 📊 Statistical Filtering (IQR)
Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1
df[col].between(Q1 - 1.5 * IQR, Q3 + 1.5 * IQR)


✅ Why:
- IQR filtering is robust to non-Gaussian distributions and avoids assumptions about symmetry.
- Applied to features like logM_star, velocityDisp, and d4000, which can have long tails due to rare galaxy types or measurement errors.
- Helps remove statistical outliers that could bias regression models or distort feature distributions.

4. 🧠 Multivariate Outlier Detection (Isolation Forest)
IsolationForest(contamination=0.02)


✅ Why:
- Captures outliers that are only apparent in the joint distribution of features (e.g. a galaxy with normal Hα but anomalous [OIII]/Hα ratio).
- Isolation Forest is non-parametric and works well with high-dimensional data.
- contamination=0.02 assumes ~2% of the data are outliers — adjustable based on survey quality.

5. 📉 PCA Visualization (Optional)
PCA(n_components=2)


✅ Why:
- Reduces dimensionality for visual inspection.
- Helps validate whether outliers flagged by Isolation Forest are truly isolated in feature space.
- Useful for debugging and understanding the structure of your dataset.

🧪 Scientific Integrity
This pipeline balances:
- Preserving rare but valid galaxies (e.g. AGNs, starbursts)
- Removing catalog artifacts and measurement errors
- Improving model performance by cleaning noisy inputs


```python
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

def clean_galaxy_data(df, contamination=0.02, plot_pca=True):
    """
    Cleans and preprocesses galaxy dataset for ML modeling by applying:
    - Initial cleaning (removing invalid values and non-galaxy objects)
    - Log transformation of emission line fluxes
    - Physical filtering based on redshift, error, and size constraints
    - Statistical outlier filtering using IQR
    - Multivariate outlier detection using Isolation Forest
    - Optional PCA visualization of detected outliers

    Parameters:
    -----------
    df : pd.DataFrame
        Raw input dataframe containing galaxy features.
    contamination : float, optional (default=0.02)
        Proportion of data expected to be outliers for Isolation Forest.
    plot_pca : bool, optional (default=True)
        Whether to display a PCA plot highlighting outliers.

    Returns:
    --------
    df_clean : pd.DataFrame
        Cleaned dataframe with outliers removed.
    """

    # --- Step 0: Remove bad values, duplicates, non-galaxies, etc. ---
    df_cleaned = df[~(df == -9999).any(axis=1)]
    df_cleaned = df_cleaned.drop_duplicates()
    df_cleaned = df_cleaned[df_cleaned['clean'] == 1]
    df_cleaned = df_cleaned[df_cleaned['type'] == 3]  # Keep only galaxies
    df_cleaned.drop(columns=['subClass', 'class', 'clean'], inplace=True)

    df = df_cleaned.copy()

    # --- Column definitions ---
    spectro_cols = [
        "h_alpha_flux", "oiii_5007_flux", "nii_6584_flux", "sii_6717_flux",
        "lick_mgb", "lick_fe5335", "d4000", "velocityDisp"
    ]
    photometry_cols = [
        "modelMag_u", "modelMag_r", "modelMag_z", "fracDev_r", "petroRad_r",
        "extinction_u", "extinction_r", "extinction_z"
    ]
    generic_cols = ["z", "zErr"]
    target_col = ["logM_star"]

    # --- Step 1: Log-transform emission line fluxes after clipping ---
    for col in ["h_alpha_flux", "oiii_5007_flux", "nii_6584_flux", "sii_6717_flux"]:
        df[col] = df[col].clip(lower=0)
        df[f"log_{col}"] = np.log1p(df[col])

    # --- Step 2: Physical filtering ---
    df = df[(df["z"] >= 0.000) & (df["z"] <= 0.30)]
    df = df[df["zErr"] < 0.01]
    df = df[df["petroRad_r"] > 0.5]
    df = df[df["fracDev_r"].between(0, 1)]
    df = df[df["extinction_u"] < 1.5]

    # --- Step 3: Statistical IQR filtering ---
    def iqr_filter(data, col):
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        return data[col].between(Q1 - 1.5 * IQR, Q3 + 1.5 * IQR)

    for col in ["logM_star", "velocityDisp", "d4000"]:
        df = df[iqr_filter(df, col)]

    # --- Step 4: Multivariate outlier detection ---
    feature_cols = [f"log_{col}" for col in ["h_alpha_flux", "oiii_5007_flux", "nii_6584_flux", "sii_6717_flux"]] \
                   + ["lick_mgb", "lick_fe5335", "d4000", "velocityDisp"] \
                   + photometry_cols + generic_cols

    df = df.dropna(subset=feature_cols)
    df_features = df[feature_cols]

    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(df_features)

    iso = IsolationForest(contamination=contamination, random_state=42)
    outlier_flags = iso.fit_predict(features_scaled)

    df_clean = df[outlier_flags == 1].copy()

    # --- Optional: PCA visualization ---
    if plot_pca:
        pca = PCA(n_components=2)
        pca_result = pca.fit_transform(features_scaled)
        plt.figure(figsize=(8, 6))
        plt.scatter(pca_result[:, 0], pca_result[:, 1],
                    c=(outlier_flags == -1), cmap='coolwarm', alpha=0.6)
        plt.title("PCA of Galaxy Features with Outliers Highlighted")
        plt.xlabel("PC1")
        plt.ylabel("PC2")
        plt.grid(True)
        plt.show()

    return df_clean
```
